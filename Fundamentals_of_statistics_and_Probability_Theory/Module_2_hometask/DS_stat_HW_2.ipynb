```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

#### Задача 1.
Сотрудники компании Shmuber, основанной в 2008 году и которая опубликовала свою отчетность, принесли Вам некоторую её часть для публикации в вашем СМИ. Вы уже "собаку съели" на похожих задачах и сразу видите, что графики на самом деле не дают явного ответа на возможные вопросы читателя отчета. 

```python
company_revenue = pd.read_csv('Task_1.csv')
company_revenue.rename(columns = {'revenue, $K' : 'revenue'}, inplace=True)
```

```python
plt.figure(figsize=(10, 7))
sns.lineplot(data=company_revenue, x='year', y='revenue', hue='company')
plt.xlabel('Год')
plt.ylabel('Доход компании, тыс. $')
plt.title('Динамика роста дохода компаний с 2008 по 2015 год', fontsize=12)
plt.show()
```

```python
total_company_revenue = company_revenue\
.groupby('company')['revenue']\
.sum()\
.reset_index()\
.sort_values('revenue', ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(data=total_company_revenue, x='company', y='revenue', alpha=0.5)
plt.title('Совокупный доход компаний за период с 2008 по 2015 год', fontsize=12)
plt.xlabel('Компания')
plt.ylabel('Совокупный доход, тыс. $')
plt.show()
```

В исходном графике демонстрировалась динамика роста отношения прибыли компаний к прибыли компании за 2008 год. Это не позволяет оценить реальный рост прибыли компаний по годам. Поэтому я решил исследовать абсолютную величину прибыли, измеряемую в тыс.$.

#### Задача 2.
Одна консалтинговая компания попросила вас опубликовать годовую статистику о результатах работы компаний на рынке бытовой техники и прислала вам диаграмму из их отчета. Вы как опытный профессионал задумались перед тем, как публиковать этот материал, как следует изменить визуализацию, чтобы она отражала реальное положение вещей?

```python
company_revenue2 = pd.read_csv('Task_2.csv')
company_revenue2.rename(columns={'revenue, $M' : 'revenue'}, inplace=True)
```

```python
plt.figure(figsize=(10, 7))
sns.barplot(data=company_revenue2.sort_values('revenue', ascending=False), x='company', y='revenue', log=True)
plt.title('Прибыль компаний в 2019 году', fontsize=12)
plt.xlabel('Компания')
plt.ylabel('Прибыль компании, млн.$')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.pie(x=company_revenue2['revenue'], labels=company_revenue2['company'])
plt.title('Доля прибыли компаний на рынке бытовой техники в 2019 году', fontsize=12)
plt.show()
```
