# A0_1.ipynb — Введение: От Matplotlib к Plotly — Практика

В теоретической части вы освежили в памяти методы работы с библиотекой Matplotlib на примере построения графика средних температур.

Подключим библиотеки и прочитаем наш набор данных.

```python
import pandas as pd
import matplotlib.pyplot as plt

import io
import requests
```

```python
df = pd.read_csv('ERA5_TA-_0002m_ADMIN-0_Glob_025d_01y_1979_2018.csv', parse_dates=True, sep=';', index_col=0)
w_africa = ('SEN','GMB','GNB','GIN','SLE','LBR')
africa = df.loc[:,w_africa]
```

В теоретической части мы получили следующий результат

```python
ax = africa.plot(title='Area averaged ERA5 temperature data, for selected countries in W.Africa.');
ax.set_xlabel("Year");
ax.set_ylabel("Celcius");
```

И вам предлагалось улучшить визуализацию этого графика:
- позиционировать легенду под графиком, в горизонтальной ориентации;
- заменить цвет фона на #EDE8D7;
- сделать заголовок двустрочным.

## Позиционируем легенду под графиком

[https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html]

Позиционировать легенду несложно. Так как необходимо поместить легенду внизу графика, воспользуемся параметром `loc='lower center'`.

```python
ax = africa.plot(title='Area averaged ERA5 temperature data, for selected countries in W.Africa.');
ax.set_xlabel("Year");
ax.set_ylabel("Celcius");
ax.legend(loc='lower center');
```

Кажется, это не то, что мы хотели. Сделаем легенду горизонтальной, для этого используем параметр `ncol=6`, так как у нас шесть элементов легенды.

```python
ax = africa.plot(title='Area averaged ERA5 temperature data, for selected countries in W.Africa.');
ax.set_xlabel("Year");
ax.set_ylabel("Celcius");
ax.legend(loc='lower center', ncol=6);
```

Этот результат уже близок желаемому, но легенда перекрывает график. Подвинем её немного вниз — под подпись на оси Х, используя параметр `bbox_to_anchor`.

```python
ax = africa.plot(title='Area averaged ERA5 temperature data, for selected countries in W.Africa.');
ax.set_xlabel("Year");
ax.set_ylabel("Celcius");
ax.legend(loc='lower center', ncol=6, bbox_to_anchor=(0.5, -0.28));
```

Вуаля, мы получили то, что хотели: позицинировали легенду под график.

## Заменим цвет фона на #EDE8D7

Это просто сделать, используя метод `set_facecolor`, передавав ему в аргумент hex код цвета как строчку.

```python
ax = africa.plot(title='Area averaged ERA5 temperature data, for selected countries in W.Africa.');
ax.set_xlabel("Year");
ax.set_ylabel("Celcius");
ax.legend(loc='lower center', ncol=6, bbox_to_anchor=(0.5, -0.28));
ax.set_facecolor('#EDE8D7')
```

## Сделаем заголовок двухстрочным

Ничего проще этого быть не может, добавим в `title` символ переноса строки `\n` .

```python
ax = africa.plot(title='Area averaged ERA5 temperature data,\n for selected countries in W.Africa.');
ax.set_xlabel("Year");
ax.set_ylabel("Celcius");
ax.legend(loc='lower center', ncol=6, bbox_to_anchor=(0.5, -0.28));
ax.set_facecolor('#EDE8D7')
```
