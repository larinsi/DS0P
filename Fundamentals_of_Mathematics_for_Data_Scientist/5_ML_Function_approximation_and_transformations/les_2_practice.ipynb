```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Задача 1.
Точки: (2; 12.258), (4; 17.24), (8; 30.151)


Функция: $f(x)=3x+5$

```python
plt.figure(figsize=(10, 7))
points_x = [2, 4, 8]
points_y = [12.258, 17.24, 30.151]

X = np.arange(0, 10, 0.01)
Y = [3*x+5 for x in X]
plt.scatter(points_x, points_y, color='red')
plt.plot(X, Y, color='blue')
plt.show()
```

```python
x = Symbol('x')
f = 3*x + 5
f
```

```python
MSE = 0
for i, y in enumerate(points_y):
    MSE += (y - f.subs(x, points_x[i]))**2
print(MSE/3)
```

### Задача 3.
Точки: (2; 4.872), (4; 29.707), (8; 246.971), (10; 485.727), (12; 840.658)


Функция: $0.5x^3-0.25x^2+0.75x+1.25$

```python
plt.figure(figsize=(10,7))
points_x = [2, 4, 8, 10, 12]
points_y = [4.872, 29.707, 246.971, 485.727, 840.658]
X = np.arange(0, 13, 0.01)
Y = [0.5*x**3 - 0.25*x**2 + 0.75*x + 1.25 for x in X]
plt.scatter(points_x, points_y, color='red')
plt.plot(X, Y, color='blue')
plt.show()
```

```python
f3 = 0.5*x**3 - 0.25*x**2 + 0.75*x + 1.25
f3
```

```python
MSE = 0
for i, y in enumerate(points_y):
    MSE += (y - f3.subs(x, points_x[i]))**2
print(MSE/5)
```
