```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

```python
f = 0.8 * x + 20 * y
f
```

```python
plot3d(f)
```

```python
f = x * y / sqrt(x**2 + y**2 - 5)
f
```

```python
plot3d(f)
```

```python
f = sin(sqrt(x ** 2 + y ** 2))
f
```

```python
plot3d(f)
```
