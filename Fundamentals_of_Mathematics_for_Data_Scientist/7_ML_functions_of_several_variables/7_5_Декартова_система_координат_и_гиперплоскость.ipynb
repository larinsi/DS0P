```python
from sympy import *
from sympy.plotting import plot, plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

```python
a0 = 3
a1 = 0.5

f = a1 * x + a0
f
```

```python
plot(f)
```

```python
a0 = 3
a1 = 0.5
a2 = 4

f = a2 * y + a1 * x + a0
f
```

```python
plot3d(f)
```
