```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

## 1. 

```python
f1 = 5 * x - 3 * y + 2
f1
```

```python
solve(f1)
```

```python
f2 = 3 / 5 * y - 2 / 5

plot(f2)
```

```python
plot3d(f1)
```
