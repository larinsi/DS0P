```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

```python
f1 = 2*x + 5*y + 1
f1
```

```python
solve(f1)
```

```python
func_x1 = -5*y/2 - 1/2
func_x1
```

```python
plot(func_x1)
```

```python
f2 = x**2 + y**2 - 1
f2
```

```python
solve(f2)
```

```python
func_x2 = -sqrt(1 - y**2)
func_x2
```

```python
plot(func_x2, (y, -1, 1))
plot(sqrt(1 - y**2), (y, -1, 1))
```

```python
f3 = x * cos(y)
f3
```

```python
solve(f3)
```
