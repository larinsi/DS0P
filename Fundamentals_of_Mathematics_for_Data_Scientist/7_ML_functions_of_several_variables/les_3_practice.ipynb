```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

```python
f1 = sin(sqrt(x**2 + y**2))
f1
```

```python
plot3d(f1)
```

```python
f2 = exp(-0.9*x**2 - 0.45*(x-y)**2)
f2
```

```python
plot3d(f2, (x, -3, 3), (y, -3, 3))
```

```python
f3 = x * cos(y)
f3
```

```python
plot3d(f3)
```
