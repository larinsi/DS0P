### Задание 2. Проведите исследование полученных функций. 

1. $5x + 1.5$
2. $2e^{0.2x}-1$
3. $3\sqrt x - 15$
4. $2\log(x) + 5$
5. $-5x^2 - 10x + 100$




```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
```

#### Функция $5x+1.5$
* Ноль в точке $x=-0.3$, 
* отрицательна при $ x < -0.3$, 
* положительна при $ x > -0.3$,
* возрастает на всей области определения функции,
* не является четной и нечетной.

```python
f1 = 5*x + 1.5
plot(f1)
```

```python
print(solve(f1), solve_univariate_inequality(f1<0, x), solve_univariate_inequality(f1>0, x))
```

#### Функция $2e^{0.2x}-1$
* Ноль в точке $x=-3.466$ ($-5ln(2)$),
* отрицательна при $x<-5ln(2)$,
* положительна при $x>-5ln(2)$,
* возрастает на всей области определения функции,
* не является четной и нечетной.

```python
f2 = 2*exp(0.2*x) - 1
plot(f2)
```

```python
print(solve(f2), solve_univariate_inequality(f2<0, x), solve_univariate_inequality(f2>0, x))
```

#### Функция $3\sqrt x - 15$
* Ноль в точке $x=25$,
* отрицательна при $0<=x<25$,
* положительна при $x>25$,
* возрастает на всей области определения функции,
* не является четной и нечетной.

```python
f3 = 3*sqrt(x) - 15
plot(f3, (x, 0, 100))
```

```python
print(solve(f3), solve_univariate_inequality(f3<0, x), solve_univariate_inequality(f3>0, x))
```

#### Функция $2log(x)+5$
* Ноль в точке $x=e^{-5/2}$,
* отрицательна при $0<x<e^{-5/2}$,
* положительна при $x>e^{-5/2}$,
* возрастает на всей области определения функции,
* не является четной и нечетной.

```python
f4 = 2*ln(x) + 5
plot(f4)
```

```python
print(solve(f4), solve_univariate_inequality(f4<0, x), solve_univariate_inequality(f4>0, x))
```

#### Функция $-5x^2-10x+100$
* Нули при $x=-1+\sqrt21$ и $x=-1-\sqrt21$,
* отрицательна при $x<-1-\sqrt21$ и $x>-1+\sqrt21$,
* положительна при $-1-\sqrt21<x<-1+\sqrt21$,
* возрастает при $x<-1$, убывает при $x>-1$,
* не является четной и нечетной.

```python
f5 = -5*x**2 - 10*x + 100
plot(f5)
```

```python
print(solve(f5), solve_univariate_inequality(f5<0, x), solve_univariate_inequality(f5>0, x))
```

```python
# Проверка функции на четность
print(f5.subs(-10, x) == f5.subs(10, x))
```
