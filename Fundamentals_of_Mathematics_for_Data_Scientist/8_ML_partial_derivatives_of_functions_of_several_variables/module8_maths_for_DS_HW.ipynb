### Задача 1. 
Найдите частные производные данных функций. Постройте графики этих производных, а также найдите их нули.


а) $f(x,y)=x(y-5)^2$

```python
x, y = symbols('x, y')
```

```python
f = x * (y - 5)**2
```

```python
diff_x = diff(f, x)
diff_x
```

```python
plot(diff_x)
```

```python
solve(diff_x)
```

```python
diff_y = diff(f, y)
diff_y
```

```python
plot3d(diff_y)
```

```python
solve(diff_y)
```

$f(x,y)=x(y-5)^2$:
* $f'_x=(y-5)^2$, ноль в точке $y=5$,
* $f'_y=x(2y-10)$, нули при $x=0$ и $y=5$.

$f(x,y)=(y-\frac{e}{20})^2\sin(\pi x)$:
* $f'_x=\pi(y-\frac{e}{20})^2cos(\pi x)$, нули при $x=\frac{1}{2}$, $x=\frac{3}{2}$, $y=\frac{e}{20}$,
* $f'_y=(2y-\frac{e}{10})sin(\pi x)$, нули при $x=0$, $x=1$, $y=\frac{e}{20}$.

$f(x,y)=x sin(x) sin(y-2)$:
* $f'_x=x sin(y-2) cos(x) + sin(x) sin(y-2)$, нули при $y=2$, $y=2+\pi$,
* $f'_y=x sin(x) cos(y-2)$, нули при $x=0$, $x=\pi$, $y=2+\frac{3\pi}{2}$, $y=\frac{\pi}{2} + 2$

Данные для аппроксимации в таблице ниже:


|масса     |мощность двигателя   |время разгона     |
|:--------:|:-------------------:|:----------------:|

|1570 кг.  |   343 л.с.          |  5.2 с.          |
|1520 кг.  |   431 л.с.          |    ? с.          |


Выпишите MSE, найдите его частные производные, затем с помощью sympy найдите точку минимума и предскажите время разгона до 100 км/ч автомобиля массой 1520 кг и с максимальной мощностью двигателя равной 431 лс.


```python
a2, a1, a0, x2, x1 = symbols('a2, a1, a0, x2, x1')
```

```python
MSE = 1/3 * ((a2*1945 + a1*560 + a0 - 4.3)**2 + \
             (a2*1495 + a1*340 + a0 - 4.9)**2 + \
             (a2*1570 + a1*343 + a0 - 5.2)**2)
MSE
```

```python
MSEa2 = diff(MSE, a2)
MSEa2
```

```python
MSEa1 = diff(MSE, a1)
MSEa1
```

```python
MSEa0 = diff(MSE, a0)
MSEa0
```

```python
nonlinsolve([MSEa2, MSEa1, MSEa0], [a2, a1, a0])
```
