```python
from sympy import *
from sympy.plotting import plot
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

### Задача 1. 
Найдите частные производные данных функций. Постройте графики этих производных, а также найдите их нули.


а) $f(x,y)=x(y-5)^2$

```python
x, y = symbols('x, y')
```

```python
f = x * (y - 5)**2
```

```python
diff_x = diff(f, x)
diff_x
```

```python
plot(diff_x)
```

```python
solve(diff_x)
```

```python
diff_y = diff(f, y)
diff_y
```

```python
plot3d(diff_y)
```

```python
solve(diff_y)
```

$f(x,y)=x(y-5)^2$:
* $f'_x=(y-5)^2$, ноль в точке $y=5$,
* $f'_y=x(2y-10)$, нули при $x=0$ и $y=5$.

б) $f(x,y)=(y-\frac{e}{20})^2\sin(\pi x)$

```python
f = (y - exp(1)/20)**2 * sin(pi*x)
```

```python
diff_x = diff(f, x)
diff_x
```

```python
plot3d(diff_x)
```

```python
solve(diff_x)
```

```python
diff_y = diff(f, y)
diff_y
```

```python
plot3d(diff_y)
```

```python
solve(diff_y)
```

$f(x,y)=(y-\frac{e}{20})^2\sin(\pi x)$:
* $f'_x=\pi(y-\frac{e}{20})^2cos(\pi x)$, нули при $x=\frac{1}{2}$, $x=\frac{3}{2}$, $y=\frac{e}{20}$,
* $f'_y=(2y-\frac{e}{10})sin(\pi x)$, нули при $x=0$, $x=1$, $y=\frac{e}{20}$.
