```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

```python
f = (y - exp(1) / 20)**2 * sin(pi * x)
f
```

```python
plot3d(f)
```

```python
solve(f)
```
