```python
# Импортируем необходимые библиотеки

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
plt.style.use('seaborn-darkgrid')
%matplotlib inline
```

```python
# Загружаем данные

df_courses = pd.read_csv('Data/courses.csv', index_col=0)
df_course_contents = pd.read_csv('Data/course_contents.csv')
df_progresses = pd.read_csv('Data/progresses.csv')
df_progress_phases = pd.read_csv('Data/progress_phases.csv', parse_dates=['start_date', 'finish_date'])
df_students = pd.read_csv('Data/students.csv', parse_dates=['birthday'])
```

```python
# Объединяем датасеты в один общий

main = df_courses.merge(df_course_contents, how='right', left_on='id', right_on='course_id')
main.drop(columns='id', inplace=True)
main = main.merge(df_progresses, how='right', on='course_id')
main.rename(columns={'id' : 'progress_id', 'title' : 'course_title', 'field' : 'course_field'}, inplace=True)
main = main.merge(df_progress_phases, on=['progress_id', 'module_number', 'lesson_number'], how='right')
main = main[['progress_id', 'module_number', 'lesson_number', 'status', 'start_date', 'finish_date',\
             'student_id', 'course_id', 'course_title', 'course_field', 'module_title', 'lesson_title',\
            'lesson_token', 'is_video', 'is_homework']]
```

```python
# Количество курсов

main.course_id.nunique()
```

```python
# Количество модулей на каждом курсе

main.groupby('course_title')['module_number'].nunique().reset_index()
```

```python
# Количество уроков на каждом модуле каждого курса

main.groupby(['course_title', 'module_number'])['lesson_number'].nunique().reset_index()
```

```python
# Медианное количество уроков в модуле на каждом курсе

main.groupby(['course_title', 'module_number'])['lesson_number'].nunique().reset_index()\
.groupby('course_title')['lesson_number'].median().reset_index()
```

```python
# Количество учеников на каждом курсе

main.groupby('course_title')['student_id'].nunique().reset_index()
```

```python
# Минимальный, максимальный, средний, медианный возраст студентов
# Очищаем данные от пропущенных и некорректных значений в столбце 'birthday'

students_age = df_students.dropna(subset=['birthday'])
students_age.loc[:,'birthday'] = pd.to_datetime(students_age.loc[:,'birthday'], errors='coerce')
students_age = students_age[students_age['birthday'].notnull()]
```

```python
# Вычисляем возраст студентов

now = pd.Timestamp('now')
students_age['age'] = (now - students_age['birthday']).astype('<m8[Y]')
# Будем считать, что возраст студентов попадает в интервал от 16 до 70 лет
students_age = students_age[(students_age['age'] >= 16) & (students_age['age'] <= 70)]
```

```python
# Подготовка датафрейма для расчета показателей возраста в разрезе курсов

main_copy = main.copy(deep=True)
main_copy = main_copy.merge(students_age[['id', 'age']], left_on='student_id', right_on='id', how='left')
main_copy.dropna(inplace=True)
course_title = list(main_copy.course_title.unique())
df_course_age = pd.DataFrame(columns=main_copy.columns)

for course in course_title:
    df_course_age = pd.concat([df_course_age, main_copy.groupby('course_title')\
                              .get_group(course).drop_duplicates(subset=['student_id'])], ignore_index=True)
```

```python
# Минимальный возраст по всем курсам

df_course_age['age'].min()
```

```python
# Максимальный возраст по всем курсам

df_course_age['age'].max()
```

```python
# Средний возраст по всем курсам

df_course_age['age'].mean()
```

```python
# Медианный возраст по всем курсам

df_course_age['age'].median()
```

```python
# Минимальный возраст студентов на каждом курсе

df_course_age.groupby('course_title')['age'].min().reset_index()
```

```python
# Максимальный возраст студентов на каждом курсе

df_course_age.groupby('course_title')['age'].max().reset_index()
```

```python
# Средний возраст студентов на каждом курсе

df_course_age.groupby('course_title')['age'].mean().reset_index()
```

```python
# Медианный возраст студентов на каждом курсе

df_course_age.groupby('course_title')['age'].median().reset_index()
```

```python
# Постройте bar-chart, отражающий количество студентов на каждом курсе. 
# Ticks нужно развернуть так, чтобы они были читаемы

data = main.groupby('course_title')['student_id'].nunique().reset_index()

plt.subplots(figsize=(15,12))
sns.barplot(data=data, x='course_title', y='student_id')
plt.xticks(rotation=90, fontsize=12)
plt.xlabel('')
plt.ylabel('Число студентов, чел.', fontsize=16)
plt.title('Число студентов на курсах', fontsize=18)

plt.show()
```

```python
# Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, 
# отражающий количество студентов на каждом курсе. График должен иметь заголовок. 
# Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, 
# к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. 
# На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. 
# Рамки у графика быть не должно ⭐

data = main.groupby(['course_title', 'course_field'])['student_id']\
.nunique()\
.reset_index()\
.sort_values('student_id', ascending=True)

def set_color(field):
    if field == 'Marketing':
        return 'blue'
    elif field == 'Design':
        return 'red'
    elif field == 'Development':
        return 'green'
    else:
        return 'orange'
    
data['color'] = data['course_field'].apply(lambda field: set_color(field))

fig, ax = plt.subplots(figsize=(16,8))
ax = plt.barh(y=data['course_title'], width=data['student_id'], color=data['color'], alpha=0.1)

plt.axvline(x=data['student_id'].median(), color='grey')

fig.suptitle('Распределение студентов по курсам', fontsize=18, y=0.93)
plt.xlabel('Количество студентов, чел.', fontsize=14)

blue_patch = mpatches.Patch(color='blue', label='Marketing', alpha=0.1)
red_patch = mpatches.Patch(color='red', label='Design', alpha=0.1)
green_patch = mpatches.Patch(color='green', label='Development', alpha=0.1)
orange_patch = mpatches.Patch(color='orange', label='Business', alpha=0.1)

plt.legend(handles=[blue_patch, red_patch, green_patch, orange_patch])

plt.show()
```

## 3. Выявление проблемных модулей

```python
# Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля 
# (разность между временем начала и окончания выполнения домашней работы) для каждого курса. 
# Если домашних заданий в модуле несколько, 
# то считать разность между временем начала выполнения первой домашней работы 
# и временем окончания выполнения последней домашней работы в модуле


df_module_time = main[(main['is_homework']) & (main['status'] == 'done')]
df_module_time = df_module_time[['course_title', 'module_number', 'lesson_number', 'start_date', 'finish_date', 'student_id']]

# Определяем время начала прохождения модуля
# Для каждого модуля устанавливаем первый урок, включающий домашнюю работу
df_start_lesson = df_module_time.groupby(['course_title', 'module_number', 'student_id'])\
['lesson_number'].min().reset_index()

# Для каждого урока определяем время начала выполнения работы
df_start_time = df_module_time.groupby(['course_title', 'module_number', 'student_id', 'lesson_number'])\
['start_date'].min().reset_index()

# Сопоставляем каждому первому уроку с домашней работой время начала выполнения домашней работы
df_start = df_start_lesson.merge(df_start_time, on=['course_title', 'module_number', 'student_id', 'lesson_number'], how='left')
df_start = df_start[['course_title', 'module_number', 'student_id', 'start_date']]

# Определяем время окончания прохождения модуля
# Для каждого модуля устанавливаем последний урок, включающий домашнюю работу
df_finish_lesson = df_module_time.groupby(['course_title', 'module_number', 'student_id'])\
['lesson_number'].max().reset_index()

# Для каждого урока определяем время окончания домашней работы
df_finish_time = df_module_time.groupby(['course_title', 'module_number', 'student_id', 'lesson_number'])\
['finish_date'].max().reset_index()

# Сопоставляем каждому последнему уроку с домашней работой время окончания работы
df_finish = df_finish_lesson.merge(df_finish_time, on=['course_title', 'module_number', 'student_id', 'lesson_number'], how='left')
df_finish = df_finish[['course_title', 'module_number', 'student_id', 'finish_date']]

# Соединяем датафреймы, чтобы получить датафрейм, в котором есть дата начала и дата окончания прохождения модуля 
df_module_time = df_start.merge(df_finish, on=['course_title', 'module_number', 'student_id'])

# Итоговый датафрейм с значениями времязатрат каждого студента на модули
df_module_time['time_spent'] = df_module_time['finish_date'] - df_module_time['start_date']

# Переводим затраченное время в дни
df_module_time['time_spent'] = df_module_time['time_spent'].apply(lambda x: x.total_seconds() / 86400)
```

```python
# Определение минимального, среднего, медианного и максимального значений затраченного на прохождение модуля времени

df_module_time.groupby(['course_title', 'module_number'])['time_spent'].describe()\
.rename(columns={'50%' : 'median'})[['min', 'mean', 'median', 'max']]
```

```python
# На основании первого пункта построить line-graph с медианным временем прохождения каждого модуля для каждого курса.
# 15 графиков. Графики должны иметь заголовки

data = df_module_time.groupby(['course_title', 'module_number'])['time_spent'].median().reset_index()

fig, axs = plt.subplots(figsize=(15,60), nrows=15, ncols=1)
plt.subplots_adjust(hspace=0.3)
for i, course in enumerate(course_titles):
    ax = axs[i]
    sns.lineplot(x=data[data['course_title'] == course]['module_number'],\
                 y=data[data['course_title'] == course]['time_spent'],\
                 ax=ax)
    ax.set_ylabel('Затраченное время, дней', fontsize=14)
    ax.set_xlabel('Номер модуля', fontsize=14)
    ax.set_title(course, fontsize=16)
    ticks = range(1, data[data['course_title'] == course].module_number.max() + 1)
    ax.set_xticks(ticks)

plt.show()
```

```python
# Чтобы выявить сезонность, посчитать медианное время выполнения домашней работы по месяцам 
# (12 месяцев, январь-декабрь) для каждого курса.

df_homework_month = main[(main['is_homework']) & (main['status'] == 'done')]\
[['start_date', 'finish_date', 'course_title']]

df_homework_month['time_spent'] = df_homework_month['finish_date'] - df_homework_month['start_date']

# Вычисляем время выполнения работы в днях
df_homework_month['time_spent'] = df_homework_month['time_spent'].apply(lambda x: x.total_seconds() / 86400)

# Будем считать, что выполнение домашней работы относится к тому месяцу, в который она была начата
df_homework_month['month'] = df_homework_month['start_date'].dt.month
df_homework_month.groupby(['course_title', 'month'])['time_spent'].median().reset_index()
```

```python
# На основании третьего пункта построить line-graph, на который будут нанесены линии для каждого курса
# с медианным временем выполнения домашней работы по месяцам. 15 линий на графике. 
# График должен иметь заголовок. Ось, отражающая время прохождения, должна быть подписана. 
# Линия для каждого курса должна иметь свой цвет (нужна легенда). Рамок у графика быть не должно ⭐


data = df_homework_month.groupby(['course_title', 'month'])['time_spent'].median()\
.reset_index().sort_values('month')

fig, ax = plt.subplots(figsize=(20,15))

sns.lineplot(data=data,
            x='month',
            y='time_spent',
            hue='course_title',
            ax=ax,
            alpha=0.5)

ax.set_title('Медианное время выполнения домашней работы в разрезе курсов', fontsize=18)
ax.set_ylabel('Время выполнения, дней', size=14)
ax.set_xlabel('')
ticks =  range(1,13)

ax.set_xticks(ticks)


plt.show()
```

```python
# Определяем количество студентов, которые завершили выполнение домашнего задания в разрезе модулей курсов

df_work_ends = main[(main['is_homework']) & (main['status'] == 'done')]\
.groupby(['course_title', 'module_number'])['student_id'].nunique().reset_index()
df_work_ends.rename(columns={'student_id' : 'num_of_completed'}, inplace=True)
```

Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

**Описание метрики успеваемости.**
Будем считать, если потраченное студентом время на прохождение модуля меньше рассчитанного, значит студент не является проблемным, в противном случае,- является.


Расчет времени для сравнения:
   1. Можно сравнивать время прохождения с медианным временем
   2. К медианному времени прохождения модуля можно прибавить некое $Х$, которое будет компенсировать определенные обстоятельства. Примем $Х=2$ и будем считать, что этого времени хватит для устранения форс-мажоров, по причине которых студент не смог придерживаться намеченного плана обучения;
   3. $Х$ - количество дней, зависящее от времени года начала выполнения задания. Как видно из вышеприведенных выкладок, сезонность оказывает влияние на время прохождения модулей.
   4. Установить значение $Х$ исходя из возраста студента. Это может сработать, потому что, с возрастом обучаемость снижается и человеку необходимо больше времени на освоение материала. Но здесь получится отслеживать показатель только по заполнившим дату рождения студентам.


Можно сделать надстройку над предыдущей метрикой чтобы сузить число неуспевающих студентов и для каждого курса определить количество модулей или конкретные модули, по которым превышение рассчитанного времени прохождения определяет студента как проблемного



```python
# Выявление проблемных студентов

df_median_time = df_module_time.groupby(['course_title', 'module_number'])['time_spent'].median().reset_index()
df_median_time.rename(columns={'time_spent' : 'median_time'}, inplace=True)
```

```python
df_performance = df_module_time.merge(df_median_time, on=['course_title', 'module_number'], how='left')
df_performance['metrics_one'] = df_performance['median_time'] < df_performance['time_spent']
df_performance['metrics_two'] = df_performance['median_time'] + 2 < df_performance['time_spent']
df_performance['month'] = df_performance['start_date'].dt.month
```

```python
def set_coef_by_year_period(month):
    if 1 <= month < 7:
        return 1
    elif 7 <= month < 10:
        return 3
    elif 10 <= month <= 12:
        return 2
    
    
df_performance['season_time'] = df_performance['month'].apply(lambda month: set_coef_by_year_period(month)) + df_performance['median_time']
df_performance['metrics_three'] = df_performance['season_time'] < df_performance['time_spent']
```

```python
df_performance = df_performance.merge(df_course_age[['student_id', 'age']], on='student_id', how='left')
```

```python
def set_coef_by_age(age):
    if age < 30:
        return 1
    elif 30 <= age < 50:
        return 2
    elif age >= 50:
        return 3

    
df_performance['age_time'] = df_performance['age'].apply(lambda age: set_coef_by_age(age)) + df_performance['median_time']
df_performance['metrics_four'] = df_performance['age_time'] < df_performance['time_spent']
```

```python
# Считаем для каждого студента количество модулей, которое он выполнил хуже, чем необходимо согласно третьей метрике 

df_performance = df_performance.merge(df_performance.groupby(['course_title', 'student_id'])['metrics_three'].sum().reset_index(),
                                      on=['course_title', 'student_id'], how='left')
df_performance.rename(columns={'metrics_three_x' : 'metrics_three', 'metrics_three_y' : 'num_violations_metric_three'}, inplace=True)
```

```python
# Считаем для каждого студента количество модулей, которое он выполнил хуже, чем необходимо согласно четвертой метрике 

df_performance = df_performance.merge(df_performance.groupby(['course_title', 'student_id'])['metrics_four'].sum().reset_index(),
                                      on=['course_title', 'student_id'], how='left')
df_performance.rename(columns={'metrics_four_x' : 'metrics_four', 'metrics_four_y' : 'num_violations_metric_four'}, inplace=True)
```
