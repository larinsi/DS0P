Постройте точечный график, на котором по оси х будет указан год, а на оси у - население. Задайте цвет каждой отметки в зависимости от величины минимальной зарплаты в конкретном штате и конкретном году.

```python
df = pd.read_csv('wage-data-coast-with-population.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,9))

sns.scatterplot(data=df, x='Year', y='Population', ax=ax, hue='Salary', size='Salary', alpha=0.5)
ax.set_title('Динамика населения США')
ax.set_xlabel('Год')
ax.set_ylabel('Население, млн. чел.')

plt.show()
```
